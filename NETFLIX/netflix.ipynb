# task 1: what is the most popular release year for movies in netflix

```python
movies_df=df[df['type']=='Movie']
```

```python
movies_df.head()
```

```python
movies_df['count']=1
popular_release=movies_df.groupby(['release_year']).sum().reset_index()[['release_year','count']]
```

```python
popular_release.sort_values(['count'],ascending=False)
#####popular release was on 2017 and 2018###############
```

# What year did Netflix add the most content to its platform?

```python
df.head()
```

```python
df['date_added']=pd.to_datetime(df['date_added'], infer_datetime_format=True)
```

```python
df.head()
```

```python
df['year_added']=df['date_added'].dt.strftime('%Y')
df.head()
```

```python
df['count']=1
year_added=df.groupby(['year_added']).sum().reset_index()[['year_added','count']]
```

```python
year_added.sort_values(['count'],ascending=False)
#####popular add was 2019###############
```

# What is the most popular month to add new content?

```python
df['month_added']=df['date_added'].dt.strftime('%m')
df.head()
```

```python

month_added=df.groupby(['month_added']).sum().reset_index()[['month_added','count']]
month_added.sort_values(['count'],ascending=False)
```
